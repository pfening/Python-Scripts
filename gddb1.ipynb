```python
import tensorflow as tf
import numpy as np
import pandas as pd
import sklearn as sk
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import Normalizer
from sklearn.cluster import KMeans
import seaborn as sb
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv('results.csv')
data = data.fillna({"TARGET_ACCOUNT": "UNKNOWN"})
data = data.fillna({"GDDB_ROLE_SPECIAL": "STANDARD"})
data.head()
```

```python
le = preprocessing.LabelEncoder()
data["UID"] = le.fit_transform(data['USERNAME'])
data["EID"] = le.fit_transform(data['EVENT_ID'])
data["TA"] = le.fit_transform(data['TARGET_ACCOUNT'])
data["TUID"] = le.fit_transform(data['TARGET_UNIQUE_ID'])
data["TG"] = le.fit_transform(data['TARGET_GENDER'])
data["ROLE"] = le.fit_transform(data['GDDB_ROLE'])
data["SROLE"] = le.fit_transform(data['GDDB_ROLE_SPECIAL'])
data["LAB"] = le.fit_transform(data['LABEL'])
```

```python
#data[["USERNAME", "UID","EVENT_ID","EID","TARGET_ACCOUNT","TA","TARGET_UNIQUE_ID","TUID","TARGET_GENDER","TG","GDDB_ROLE","ROLE","GDDB_ROLE_SPECIAL","SROLE","LABEL","LAB"]]
```

```python
df=data[["UID","EID","TA","TUID","TG","ROLE","SROLE","LAB"]]
norm = preprocessing.Normalizer()
X_train_norm = norm.fit_transform(df)
X_train_norm
```

```python
X = df[["UID","EID","TA","TUID","TG","ROLE","SROLE"]]
y = df["LAB"]
```

```python
X
```

```python
uid = tf.feature_column.numeric_column('UID')
eid = tf.feature_column.numeric_column('EID')
ta = tf.feature_column.numeric_column('TA')
tuid = tf.feature_column.numeric_column('TUID')
tg = tf.feature_column.numeric_column('TG')
role = tf.feature_column.numeric_column('ROLE')
srole = tf.feature_column.numeric_column('SROLE')
```

```python
feat_cols = [uid,eid,ta,tuid,tg,role,srole]
```

```python
x_train, x_eval, y_train, y_eval = train_test_split(X,y,test_size=0.3, random_state = 101)

print(x_train.shape)
print(y_train.shape)

print(x_eval.shape)
print(y_eval.shape)
```

```python
input_func = tf.estimator.inputs.pandas_input_fn(x=x_train,y=y_train,batch_size=10,num_epochs=20000,shuffle=True)
```

```python
model = tf.estimator.LinearClassifier(feature_columns=feat_cols,n_classes=2)
```

```python
model.train(input_fn=input_func,steps=20000)
```

```python
eval_input_func = tf.estimator.inputs.pandas_input_fn(x=x_eval,y=y_eval,batch_size=10,num_epochs=1,shuffle=False)
```

```python
results = model.evaluate(eval_input_func)
```
